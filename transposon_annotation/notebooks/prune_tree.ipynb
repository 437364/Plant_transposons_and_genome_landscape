Prunes a published angiosperm phylogeny (Angiosperms_100trees.tre) to the study species set using ete3. Swaps some clades for visual consistency. Renames Chamealirium japonicum for consistency with Kuo et al. papers.
Exports data/pruned_tree.tre and data/leaf_names.txt.
Input: Published tree, accessions list
Output: data/pruned_tree.tre, data/leaf_names.txt

```python
#/storage/plzen1/home/kratka/.conda/envs/ete3
import ete3
import pandas as pd
```

```python
sample_list_path = "data/accessions_genome.tsv"
df = pd.read_csv(sample_list_path, sep="\t")
samples_list = df["Sample_id"].tolist()
```

```python
tree = ete3.Tree("data/one_tree.tre", format=1)
```

```python
# replace Chionographis japonica with Chamaelirium japonicum
for leaf in tree.iter_leaves():
    if leaf.name == "Chamaelirium_japonicum":
        leaf.name = "Chionographis_japonica"
tree.prune(samples_list, preserve_branch_length=True)
```

```python
# swap Luzula and Juncus
# Node containing specific leaves
target_leaves = {"Luzula_sylvatica", "Juncus_effusus"}
node = tree.get_common_ancestor(target_leaves)
node.swap_children()
```

```python
# swap Cuscutas
# Node containing specific leaves
target_leaves = {"Cuscuta_campestris", "Cuscuta_epithymum"}
node = tree.get_common_ancestor(target_leaves)
node.swap_children()
```

```python
tree.write(outfile="data/pruned_tree.tre")
```

```python
leaf_names = tree.get_leaf_names()
print(leaf_names)
# save the list of leaf names
with open("data/leaf_names.txt", "w") as f:
    for name in leaf_names:
        f.write(name + "\n")
```
